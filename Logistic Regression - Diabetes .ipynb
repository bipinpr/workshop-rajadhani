```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diabetes = pd.read_csv('Diabetes.csv')
```

```python
diabetes.isnull().sum()
```

### Data Manipulation

```python
diabetes_df = diabetes.drop(['Pregnancies', 'Outcome'], axis = 1)
```

```python
diabetes_df
```

```python
diabetes_df.replace(0, np.nan, inplace=True)
```

```python
diabetes_df
```

```python
diabetes_df.isnull().sum()
```

### Simple Imputer

* It replaces np.nan (missing values with mean/median/mode)

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
diabetes_array = imputer.fit_transform(diabetes_df)
```

```python
diabetes_array
```

```python
diabetes_df_new = pd.DataFrame(diabetes_array, columns= diabetes_df.columns)
```

```python
diabetes_df_new
```

```python
diabetes_df_new1 = pd.concat([diabetes_df_new, diabetes['Pregnancies']], axis=1)
```

```python
diabetes_df_new1
```

```python
y = diabetes['Outcome']
```

```python
y
```

```python
X=diabetes_df_new1
```

```python
X
```

### Train/Test Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2,random_state = 44)
```

```python
X_train
```

### Logistic Algorithm

```python
from sklearn.linear_model import LogisticRegression
# penalty = to avoid overfitting
logit_reg = LogisticRegression() 
```

```python
logit_reg.fit(X_train, Y_train) 
```

```python
yhat = logit_reg.predict(X_test)
```

```python
## Predictions for Test Data
yhat
```

```python
Y_test
```

### Classification Report 

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, yhat)
```

```python
sns.heatmap(confusion_matrix(Y_test, yhat), annot=True)
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test, yhat))
```

```python
logit_reg.predict([[51.0, 72.0, 35.0, 101.0, 33.6, 0.627, 32.0, 9]])
```

```python
from sklearn .metrics import roc_auc_score
roc_auc_score(Y_test, yhat)
```
