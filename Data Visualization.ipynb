```python
#Matplotlib

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Line chart
no_of_customers=[10,20,30,45,65]
sales=[100,200,150,500,120]

plt.plot(no_of_customers,sales)
```

```python
no_of_customers=[10,20,30,45,65]
sales=[100,200,150,500,120]

plt.figure(figsize=(7,3))
plt.plot(no_of_customers,sales)

plt.xlabel("number of customers")
plt.ylabel("sales")
plt.title('Line plot')
plt.legend(["Line1"])
plt.grid()
```

```python
no_of_customers=[10,20,30,45,65]
sales_usual=[100,200,150,500,120]
sales_discounted=[150,220,190,510,620]

plt.subplot(1,2,1)
plt.plot(no_of_customers,sales_usual)
plt.xlabel("number of customers")
plt.ylabel("sales_usual")
plt.title('Line plot')
plt.grid()

plt.subplot(1,2,2)
plt.plot(no_of_customers,sales_discounted)
plt.xlabel("number of customers")
plt.ylabel("sales_discounted")
plt.title('Line plot')
plt.grid()
```

```python
no_of_customers=[10,20,30,45,65]
sales_usual=[100,200,150,500,120]
sales_discounted=[150,220,190,510,620]

plt.plot(no_of_customers,sales_usual)
plt.plot(no_of_customers,sales_discounted)
plt.xlabel("number of customers")
plt.ylabel("sales")
plt.legend(["sales_usual","sales_discounted"])
plt.title('Line plot')
plt.grid()
plt.savefig("line plot.png")

```
