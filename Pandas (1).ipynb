### Data Manipulation
To make the dataset more readable and organized

```python
#indexing by label
# loc[row , column]
# loc[ start_row:end_row , start_col : end_col]
```

```python
df.loc[:,:]
```

```python
df.loc[:,"price":"bathrooms"]
```

```python
# extract from gender col to tenure but only rows from 100 - 120 -- loc
df.loc[100:105,"price":"bedrooms"]
```

```python
# Indexing by position --> using iloc[:,:]
df.iloc[100:106,2:4]
```

```python
df.iloc[100:106,2]
```

```python
#Setting a single value to columns
df['bathrooms']=2
df
```

```python
d= pd.DataFrame({
    'name': ['John', 'Mary', 'Peter', 'Jeff', 'Bill'],
    'age': [25, 30, 45, 22, 35],
    'salary': [50000, 80000, 20000, 30000, 70000]})
d
```

```python
df3=d.sort_values(by='age',ascending=False)

df3
```

```python
#Filtering
filter1=(df3['age']>24) & (df3['salary']<60000)

filtered_df=df3[filter1]
filtered_df
```

```python
filter2=df3['salary']>60000
```

```python
filtered_2=df3[filter2]
filtered_2
```

```python

df5 = pd.DataFrame({
    'name': ['John', 'Mary', 'Peter', 'Jeff', 'Bill'],
    'age': [25, 30, 45, 22, 35],
    'salary': [50000, 80000,np.nan, 30000, 70000]})
df5
```
